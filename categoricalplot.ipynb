#### Categorical plot --> it is ised for qualitative form of data like gender name, etc.
main type of catgorical plots: 
```text
*boxplot
*violinplot
*stripplot
*swarmplot
*barplot
*countplot
```

````python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
````

````python
df = sns.load_dataset('tips')
df
````

````python
# countplot to see a paticular count
sns.countplot(x=df['sex'],hue=df['smoker']) #syntax sns.countplot(either x or either y,hue = you can set on which basis you want to sort)
````

````python
#barplot
sns.barplot(x=df['sex'],y=df['total_bill'],estimator=sum)
#here in barplot you provide quantitative data on x axis and qualitative on y axis and then you can provide estimator by default it compare them on mean but you can 
#change them according to yout own len.count.sum.min.max etc..
````

````python
#boxplot
#outliers are usually a data that stand outs either it is small or it is large but it is different from the data in that particular field 
#for eg: [66,68,76,69,77,99] --> here 99 is very far away so it is an outlier
sns.boxplot(x='day',y='tip',data=df)
````

````python
#violinplot
sns.violinplot(x='day',y='tip',data=df)
````

````python
#stripplot
sns.stripplot(x='day',y='tip',data=df)
````

````python
#swarmplot
#it shows a lil spreaded data where there is more variation of data
sns.swarmplot(x='day',y='tip',data=df)
````

````python
### combining violin plot and swarmplot together.
#writting them together combines these two plots
sns.violinplot(x='tip',y='day',data=df)
sns.swarmplot(x='tip',y='day',data=df)
````
