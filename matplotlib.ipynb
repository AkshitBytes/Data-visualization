### Learning data visualization using matplotlib seaborn plotly and cufflinks

````python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
````

### Matplotlib Basics

````python
x = np.linspace(0,5,11) # linspace means linear space it meanse between 0 to 5 give me 11 elements
y = x**2
````

````python
plt.plot(x,y) #to plot a graph matplotlib by default plots a line graph and the syntax is plt.plot(x_axis,y_axis)
plt.title("Title") #gives title to your graph
plt.xlabel("X axis")#gives a lable to your X axis
plt.ylabel("Y axis")#give a lable to your Y axis
````

### Subplots in Matplotlib


````python
plt.subplot(2,2,1)
plt.title("SUBPLOT-1") # creates subplot divide the canvas into rows and colums and then you can acces them one by one syntax:
#plt.subplot(row,column.subplot to access) if row =2 col = 2 then there are four subplots you can access 1 2 3 4 
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title("SUBPLOT-2")
plt.subplot(2,2,3)
plt.title("SUBPLOT-3")
plt.plot(y,x)
plt.subplot(2,2,4)
plt.title("SUBPLOT-4")
#to create a plot in out of these 4 subplots just use plt.plot under the subplot you want to create a plot in
````

### Object oriented way
```text
Matplotlib allows the aspect ratio, DPI and figure size to be specified when the figure object is created. You can use the figsize and dpi
Keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dot-per-inch(pixel per inch)
```

````python
fig = plt.figure(figsize=(6,8),dpi=100)
# Higher DPI → More pixels → Sharper image  
axis_1=fig.add_axes([0.1,0.1,1,1])#([left,bottom,width,height]) and all the values should be betwwwn 0 and 1 
axis_1.plot(x,y)
axis_2=fig.add_axes([0.4,0.4,0.3,0.3])
axis_2.plot(x,y)
## using object oriented way we have full control over our axis 
````
