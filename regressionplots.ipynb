````python
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
````

````python
tips = sns.load_dataset('tips')
tips
````

````python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',markers=['o','v'],scatter_kws={'s':100})#the prediction you want to make goes on y axis and therough whcih you are predicting goes on to the x axis here we are predicting how much 
# tip will be given o the basis of total_bill
# markers will change the ovals here in the graph
````
